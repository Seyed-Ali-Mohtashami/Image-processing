```python
import cv2 as cv
import matplotlib.pyplot as plt

image=cv.imread('sam.jpg')
image=cv.cvtColor(image,cv.COLOR_BGR2RGB)


plt.figure(figsize=(20,20))
plt.subplot(331)
plt.imshow(image)
plt.title('original')

row,col,tab=image.shape

#------------------------------------- flip vertical -------------------------------------
fv_image=image.copy()
for i in range(row):
    fv_image[i,:,:]=image[row-i-1,:,:]
plt.subplot(332)
plt.imshow(fv_image)
plt.title('flip vertical')

#------------------------------------ flip horizontal ------------------------------------
fh_image=image.copy()
for i in range(col):
    fh_image[:,i,]=image[:,col-i-1,:]
plt.subplot(333)
plt.imshow(fh_image)
plt.title('flip horizontal')

plt.show()
```
