```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image=cv.imread('sam.jpg',0)
plt.figure(figsize=(20,20))

plt.subplot(221)
plt.imshow(image,cmap='gray')
plt.title('original')

# -------------------------- define normalization image --------------------------
def im_norm(img):
    return img/255

# ---------------------------- define image histogram ----------------------------
def im_hist(img,value):
    r,c=img.shape
    count=0
    for i in range(r) :
        for j in range(c):
            if img[i,j]==value:
                count=count+1
    return count

# ---------------------------- make histogram diagram ----------------------------
n_image=im_norm(image)
plt.subplot(222)
plt.imshow(n_image,cmap='gray')
plt.title('notmalize')

row,col=n_image.shape
hist=np.arange(0,256,1)
for w in hist:
    hist[w]=im_hist(image,w)

x=np.arange(0,256,1)
plt.subplot(223)
plt.plot(x,hist)
plt.xticks([0,50,100,150,200,250])
plt.yticks([])
plt.title('histogram,define')
# ---------------------------- builtin histogram diagram ----------------------------
hist_builtin=cv.calcHist([image],[0],None,[256],[0,256])
plt.subplot(224)
plt.plot(hist_builtin)
plt.xticks([0,50,100,150,200,250])
plt.yticks([])
plt.title('histogram,builtin')
        
plt.show()
```
