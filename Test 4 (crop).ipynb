```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image=cv.imread('sam.jpg')
image=cv.cvtColor(image,cv.COLOR_BGR2RGB)

plt.figure(figsize=(20,20))
plt.subplot(221)
plt.imshow(image)
plt.title('original')

row,col,tab=image.shape
#------------------------------------- crop 100px from center -------------------------------------
crop_size=200
row_start=int(round((row-crop_size)/2,0))
row_end=row_start+crop_size

col_start=int(round((col-crop_size)/2,0))
col_end=col_start+crop_size

c_image=image[row_start:row_end,col_start:col_end,:]
plt.subplot(222)
plt.imshow(c_image)
plt.title('cropeed')


plt.show()
```
