```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image=cv.imread('exam2.png',0)
hist=cv.calcHist(image,[0],None,[256],[0,255])
row,col=image.shape
plt.figure(figsize=(15,15))
plt.subplot(331)
plt.imshow(image,cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('original image')

plt.subplot(332)
plt.plot(hist)
plt.title('histogram of original image')

#---------------------------------------- histogram bar by 10 point ----------------------------------------
#     cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
hist_10=cv.calcHist(image,    [0]  , None,   [26] , [0,255])
plt.subplot(334)
x=np.arange(0,255,10)
y=hist_10[:,0]
#plt.bar(x, height, width, bottom, align)
plt.bar(x,y,9)
plt.title('bar histogram of original image')

#---------------------------------------- histogram bar by 10 point ----------------------------------------
#     cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])
hist_10=cv.calcHist(image,    [0]  , None,   [26] , [0,255])
plt.subplot(335)
x=np.arange(0,255,10)
y=hist_10[:,0]

#stem([x,] y, linefmt=None, markerfmt=None, basefmt=None)
plt.stem(x,y)
plt.title('stem histogram of original image')

#-----------------------------------------histogram fill by 10 point ---------------------------------------
plt.subplot(336)
x=np.arange(0,256,1)
y=hist[:,0]
plt.fill(x,y)
plt.title('fill histogram of original image')

plt.show()
```
