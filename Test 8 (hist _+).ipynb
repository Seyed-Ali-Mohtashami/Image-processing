```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image=cv.imread('sam.jpg',0)
hist=cv.calcHist(image,[0],None,[256],[0,255])

plt.figure(figsize=(15,15))
plt.subplot(331)
plt.imshow(image,cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('original image')

plt.subplot(332)
plt.plot(hist)
plt.title('histogram of original image')

#----------------------------------------- hist + 50 -----------------------------------------
hist_plus=hist+50
plt.subplot(334)
plt.plot(hist_plus)
plt.title('histogram +50')
plt.yticks([0,50,100,150,200,250])

#----------------------------------------- hist * 50 -----------------------------------------
hist_mul=hist*50
plt.subplot(335)
plt.plot(hist_mul)
plt.title('histogram *50')
#plt.yticks([0,250,500,150,200,250])


plt.show()
```
