```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image=cv.imread('exam.png')
image=cv.cvtColor(image,cv.COLOR_BGR2GRAY)

plt.figure(figsize=(20,20))
plt.subplot(441)
plt.imshow(image,cmap='gray')
plt.title('original')
plt.xticks([])
plt.yticks([])

hist_org=cv.calcHist([image],[0],None,[256],[0,255])
plt.subplot(445)
plt.plot(hist_org)
plt.title('histogram of original image')
plt.xticks([])
plt.yticks([])

row,col=image.shape
#-------------------------------- enhamcement (cross=mean) --------------------------------
image=image/255
new_image=image.copy()
cross=np.mean(image)
print(cross)

for i in range(row):
    for j in range(col):
        if (image[i,j] >= cross):
            new_image[i,j]=image[i,j]*2
        else:
            new_image[i,j]=image[i,j]*1.2
       
plt.subplot(442)
plt.imshow(new_image,cmap='gray')
plt.title('new image(cross = mean)')
plt.xticks([])
plt.yticks([])


new_image=new_image*255
hist_new=cv.calcHist(np.uint8([new_image]),[0],None,[256],[0,255])
plt.subplot(446)
plt.plot(hist_new)
plt.title('histogram of original image')
plt.xticks([])
plt.yticks([])

#-------------------------------- enhamcement (cross=median) --------------------------------
new_image2=image.copy()
cross2=np.median(image)
print(cross2)

for i in range(row):
    for j in range(col):
        if (image[i,j] >= cross):
            new_image2[i,j]=image[i,j]*2
        else:
            new_image2[i,j]=image[i,j]*1.2
       
plt.subplot(443)
plt.imshow(new_image2,cmap='gray')
plt.title('new image(cross = median)')
plt.xticks([])
plt.yticks([])


new_image2=new_image2*255
hist_new2=cv.calcHist(np.uint8([new_image2]),[0],None,[256],[0,255])
plt.subplot(447)
plt.plot(hist_new2)
plt.title('histogram of original image')
plt.xticks([])
plt.yticks([])

#-------------------------------- enhamcement (cross= 0.25) --------------------------------
new_image3=image.copy()
cross3=0.25
print(cross3)

for i in range(row):
    for j in range(col):
        if (image[i,j] >= cross):
            new_image3[i,j]=(image[i,j]+0.2)*2
        else:
            new_image3[i,j]=image[i,j]*1.2
       
plt.subplot(444)
plt.imshow(new_image3,cmap='gray')
plt.title('new image(cross = 0.25)')
plt.xticks([])
plt.yticks([])


new_image3=new_image3*255
hist_new3=cv.calcHist(np.uint8([new_image3]),[0],None,[256],[0,255])
plt.subplot(448)
plt.plot(hist_new3)
plt.title('histogram of original image')
plt.xticks([])
plt.yticks([])

plt.show()
```
