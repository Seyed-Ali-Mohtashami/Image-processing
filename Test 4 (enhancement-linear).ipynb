```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

image=cv.imread('exam.png')
image=cv.cvtColor(image,cv.COLOR_BGR2GRAY)

#plt.figure(figsize=(20,20))
plt.subplot(221)
plt.imshow(image,cmap='gray')
plt.title('original')
plt.xticks([])
plt.yticks([])

hist_org=cv.calcHist([image],[0],None,[256],[0,255])
plt.subplot(223)
plt.plot(hist_org)
plt.title('histogram of original image')
plt.xticks([])
plt.yticks([])

row,col=image.shape
#-------------------------------- enhamcement (cross=mean , s=r*1.5 or 1.2) --------------------------------
image=image/255
new_image=image.copy()
cross=np.mean(image)
print(cross)

for i in range(row):
    for j in range(col):
        if (image[i,j] >= cross):
            new_image[i,j]=image[i,j]*2
        else:
            new_image[i,j]=image[i,j]*1.2
       
plt.subplot(222)
plt.imshow(new_image,cmap='gray')
plt.title('new image')
plt.xticks([])
plt.yticks([])


new_image=new_image*255
hist_new=cv.calcHist(np.uint8([new_image]),[0],None,[256],[0,255])
plt.subplot(224)
plt.plot(hist_new)
plt.title('histogram of original image')
plt.xticks([])
plt.yticks([])

plt.show()
```
