```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,20))
#--------------------------------------------- original image ---------------------------------------------
image = cv2.imread('sam.jpg')
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.subplot(331)
plt.imshow(img)
plt.title('Original')
#--------------------------------------------- grayscale image ---------------------------------------------
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.subplot(332)
plt.imshow(img_gray,cmap='gray')
plt.title('Grayscale')
#--------------------------------------------- make new image ---------------------------------------------
#print(img.shape)
r,g,b=cv2.split(img)

#print(r.shape)
#print(g.shape)
#print(b.shape)

img_copy=cv2.merge((r,g,b))
plt.subplot(337)
plt.imshow(img_copy)
plt.title('merge r+g+b')

plt.subplot(334)
plt.imshow(r,cmap='Reds')
plt.title('red channel')

plt.subplot(335)
plt.imshow(g,cmap='Greens')
plt.title('green channel')

plt.subplot(336)
plt.imshow(b,cmap='Blues')
plt.title('blue channel')

plt.show()
```
